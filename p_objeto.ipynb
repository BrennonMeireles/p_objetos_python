um exemplo pratico

```python
# um exemplo prático de criação de uma classe

class pessoa: # criando a classe pessoa
  def __init__(self, nome: str, idade: int, altura: float): # criando uma fuunção com os paremetros :nome, idade, altura...
    self.nome = nome     # definição das variaveis
    self.idade = idade   # .
    self.altura = altura # .

  def dizer_ola(self): # criando a função "ola"
    print(f'Olá meu nome é {self.nome}. Tenho {self.idade} anos e minha altura é {self.altura}') # exibe os parametros

  def cozinhar(sel, receita: str): # criando a função receita, que mostra oque a pessoa esta cozinhando
    print(f'Estou cozinhando um(a) {receita}!')

  def andar(self, distancia: float):
    print(f'Sai para andar. Volto quando completar {distancia} metros....')
```

```python
# criar uma instância da classe pessoa

um_mano = pessoa(nome = 'João',idade = 34, altura = 1.67)
```

```python
# Invocando os métodos de Pessoa

um_mano.dizer_ola()
um_mano.cozinhar('Estrogonoffe')
um_mano.andar(1500.9)
```

## Exemplo de código que cria uma classe

```python
# Exemplo de um código que cria uma classe

class Televisao():            # criamos a classe
    def __init__(self):       # um método especial chamado
        self.ligada = False   # um atributo do objeto
        self.canal = 2        # outro atributo do objeto
```

```python
# @title Exemplo de criação de uma instancia - cópia do objeto

tv = Televisao()  # criando a instancia de Televisão...
```

```python
# @title Exemplo de exibição do atributo do objeto

tv.canal # exibe e devolve o valor de tv
```

```python
# Exemplo de exibição do atributo do objeto

tv.ligada # exibe o valor de tv ligada
```

```python
# @title Outro exemplo de ciração de objeto

tv_sala = Televisao()
```

```python
tv_sala.ligada  # chamando o atributo
```

```python
tv_sala.canal
```

```python
# @title Exemplo de auteração de valor de atributo

tv_sala.ligada = True
tv_sala.canal = 4

tv_sala.ligada
```

```python
# @title Associar um novo comportamento

class Televisao():            # Criamos a classe
    def __init__(self):       # um método especial chamado contrutor
        self.ligada = False   # um atributo do objeto
        self.canal = 2        # Outro atributo do objeto

    def muda_canal_para_baixo(self):  # um novo método de televisão
        self.canal -= 1
    def muda_canal_para_cima(self):   # outro novo método de televisão
        self.canal += 1
```

### Um problema na nossa classe televisão é que não encontramos os limites dos canais
###  vamos resolver isso modificando o construtor da classe

```python

# @ title Modificar o contrutor

class Televisao():
    def __init__(self, min, max):
        self.ligada = False       # um atributo do objeto
        self.canal = 2            # Outro atributo do objeto
        self.cmin = min
        self.cmax = max

    def muda_canal_para_baixo(self):
        if self.canal - 1 >= self.cmin: # testa se dentro do limite
            self.canal -= 1
            print('Canal alterado!')
        else:
          print('Não é possível acessar essa faixa de canais!')

    def muda_canal_para_cima(self):
        if self.canal + 1 <= self.cmax: # testa se dentro do limite
            self.canal += 1
            print('Canal alterado')
        else:
          print('Não é possível acessar essa faixa de canais!')
```

```python
tv_quarto = Televisao(0, 30)
```

```python
tv_quarto.canal
```

```python
tv_quarto.muda_canal_para_cima()
print(f'O canal de Tv que esta passando é o {tv_quarto.canal}')
```

```python
tv_quarto.muda_canal_para_baixo()
print(f'O canal de Tv que esta passando é o {tv_quarto.canal}')
```

```python
tv_quarto.muda_canal_para_baixo()
```
